```python
import pandas as pd
```

```python
df = pd.read_csv(r'D:\program1\Projects\Lemtuznicova news\Tinkoff_api\corr_data.csv')
df
```

```python
df = df[df['week1_close'].notna()]
df.reset_index(inplace=True)
```

```python
def create_up_column(x):
    start_value = x['week1_open']
    week1_max = x['week1_high']
    week2_max = x['week2_high']
    week3_max = x['week3_high']
    week4_max = x['week4_high']
    
    res = []

    for st, wm1, wm2, wm3, wm4 in zip(start_value, week1_max, week2_max, week3_max, week4_max):
        val = max(wm1, wm2, wm3, wm4)
        if (val - st) / st * 100 > 8:
            res.append(1)
        else:
            res.append(0)

    return pd.Series(res)
```

```python
def create_down_column(x):
    start_value = x['week1_open']
    week1_low = x['week1_low']
    week2_low = x['week2_low']
    week3_low = x['week3_low']
    week4_low = x['week4_low']
    
    res = []

    for st, wm1, wm2, wm3, wm4 in zip(start_value, week1_low, week2_low, week3_low, week4_low):
        val = min(wm1, wm2, wm3, wm4)
        if (val - st) / st * 100 < -8:
            res.append(1)
        else:
            res.append(0)

    return pd.Series(res)
```

```python
new_df = df.assign(up_class=create_up_column)
new_df['up_class'].value_counts()
```

```python
# new_df = df.assign(down_class=create_down_column)
# new_df['down_class'].value_counts()
```

```python
import re

res_content = []
class_ls = new_df['up_class']

for index, text in enumerate(new_df['content']):
    reg_persent = re.compile(r'\+[\d\.]*%')
    reg_persent_pm = re.compile(r'[+-]?[\d\.]*%')
    all_number_reg = re.compile(r'\s([\+\-]*\d+\.\d+)|([\+\-]*\d+)\s')

    # if len(reg_persent.findall(text)) > 0:
    #     class_ls[index] = 1

    text = reg_persent.sub(r'', text)
    text = reg_persent_pm.sub(r'', text)
    text = all_number_reg.sub(r'', text)
    res_content.append(text)
```

```python
d = {'content': res_content, 'up_class': class_ls}
res_df = pd.DataFrame(data=d)
```
